```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
breast_cancer = datasets.load_breast_cancer()
X_bc = breast_cancer.data
y_bc = breast_cancer.target
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_bc[:, 0], X_bc[:, 1], c=y_bc, cmap='coolwarm')
plt.title('Scatter Plot of Data before PCA')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_bc)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_bc, cmap='coolwarm')
plt.title('Scatter Plot of Data after PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_pca, y_bc, test_size=0.3, random_state=42)
```

```python
classifier = RandomForestClassifier(n_estimators=100, random_state=42)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
report = classification_report(y_test, y_pred)
print(report)
```

```python
# Plotting the predictions
plt.figure(figsize=(8, 6))
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_pred, cmap='coolwarm')
plt.title('Predictions on Test Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
